```python
import numpy as np
import pandas as pd
import ast
```

```python
movies = pd.read_csv('movie_dataset/tmdb_5000_movies.csv')
credits = pd.read_csv('movie_dataset/tmdb_5000_credits.csv') 
```

```python
movies.head(2)
```

```python
movies.shape
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head() 
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name']) 
    return L 
```

```python
movies.dropna(inplace=True)
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(convert)
movies.head()
```

```python
movies['cast'] = movies['cast'].apply(lambda x:x[0:3])
```

```python
def fetch_director(text):
    Director = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            Director.append(i['name'])
    return Director
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.sample(5)
```

```python
def strip_spaces(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(strip_spaces)
movies['crew'] = movies['crew'].apply(strip_spaces)
movies['genres'] = movies['genres'].apply(strip_spaces)
movies['keywords'] = movies['keywords'].apply(strip_spaces)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python

movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
new_movie_table = movies.drop(columns=['overview','genres','keywords','cast','crew'])
```

```python
new_movie_table['tags'] = new_movie_table['tags'].apply(lambda x: " ".join(x))
new_movie_table.head()
```

```python
#Starting the vectorization
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectorized_data = cv.fit_transform(new_movie_table['tags']).toarray()
```

```python
vectorized_data.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectorized_data)
```

```python
similarity
```

```python
#Try
new_movie_table[new_movie_table['title'] == 'The Lego Movie'].index[0]
```

```python
def recommend_movie(movie):
    index = new_movie_table[new_movie_table['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_movie_table.iloc[i[0]].title)
```

```python
recommend_movie('Gandhi')
```
